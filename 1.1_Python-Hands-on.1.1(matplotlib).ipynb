```python
import matplotlib.pyplot as plt
# plt.plot(list(range(1,10)), list( i**2 for i in range(1,10)))
plt.plot(list(range(1,10)), list( i**2 for i in range(1,10)),'ro')
plt.xlabel("Indices")
plt.ylabel("Squares")
plt.title("x**2 graph")
plt.grid()
plt.show()
```

```python
import numpy as np
a = np.arange(1,4,0.2) 
print(a)
plt.plot(a, a**2,'b--', label='^2')
plt.plot(a, a**2.2,'rs', label='^2.2')
plt.plot(a, a**2.4,'g^', label='^2.4')
plt.legend()
plt.show()
```

```python
a = list(range(1,10))
print(a)
plt.plot(a, [i**2 for i in a],'b--', label='^2')
plt.plot(a, [i**2.2 for i in a],'rs', label='^2.2')
plt.plot(a, [i**2.4 for i in a],'g^', label='^2.4')
plt.legend()
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]
lines = plt.plot(x1, y1, x2, y2)

# use keyword args
plt.setp(lines[0], color='r', linewidth=2.0)

# or MATLAB style string value pairs
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)

plt.grid()
```

```python
'''working with multiple figures and axes'''
a = list(range(1,10))

t1 = [i**2 for i in a]
t2 = [i**2.2 for i in a]
t3 = [i**2.4 for i in a]

plt.figure(1)
plt.subplot(131)
plt.plot(a, t1,'bo', label='^2',)
plt.legend()
plt.grid()

plt.figure(1)
plt.subplot(132)
plt.plot(a, t2,'rs', label='^2.2')
plt.legend()

plt.figure(1)
plt.subplot(133)
plt.plot(a, t3,'g^', label='^2.4')
plt.legend()

plt.figure(2)
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]
lines = plt.plot(x1, y1, x2, y2)
# use keyword args
plt.setp(lines[0], color='r', linewidth=2.0)
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)

plt.figure(1)
plt.subplot(132)
plt.title("Figure One")

plt.figure(2)
plt.title("Figure Two")

plt.show()
```

```python
grid = plt.GridSpec(2, 2, wspace=0.4, hspace=0.3)
plt.subplot(grid[0, 0:1])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :])
# plt.subplot(grid[1, 2]);
plt.show()
```
